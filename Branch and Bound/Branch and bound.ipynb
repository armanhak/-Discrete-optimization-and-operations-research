# Max clique. Branch and Bound

```python
from my_module import CplexSolver, BnBSolver
import pandas as pd
from tqdm.notebook import tqdm
with open('file_names.txt', 'r') as f:
    files = [path.strip() for path in f.readlines()]
res = {'graph':[], 'clique':[], 'time (sec)':[], 'Timeout':[], 'Timeout (sec)':[]}
```

```python
# %%time    
for path in tqdm(files): 
    print(path)
    cplex_solver = CplexSolver(path)
    bnb = BnBSolver(graph = cplex_solver.graph, 
                    solver=cplex_solver.solver,
                    time_limit=7200
                   )
    r = bnb.run_bnb()
    res['graph'].append(path.replace('DIMACS_all_ascii/', ''))
    res['clique'].append(bnb.MAX_CLIQUE)
    res['time (sec)'].append(round(bnb.work_time,3))
    res['Timeout'].append(bnb.work_time>bnb.time_limit)
    res['Timeout (sec)'].append(bnb.time_limit)    
```

```python
result = pd.DataFrame(res)
result
```
